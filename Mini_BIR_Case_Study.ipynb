# Individual vs Corporate

```
labels_4g = ['Individual','Corporate']
values_4g = df_enquiry['Customer Type'].value_counts().values
fig1, ax1 = plt.subplots()
ax1.pie(values_4g, labels=labels_4g, shadow=True, startangle=90, autopct='%1.1f%%');
```

97.6% of customers are individuals and 2.4% belong to corporate.

# Retain Sale

```
df_retain = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Retail Sale Register.xlsx')
df_retain.head()
```

```
df_retain['COUNT'].value_counts()
```

```
df_retain = df_retain.where(df_retain['COUNT'] != -1)
```

```
plt.figure(figsize=(15,8));
g = sns.countplot(df_retain['Model Group']);
plt.xticks(rotation=45);
```

```
plt.figure(figsize=(20,10));
g = sns.countplot(df_retain['Model Group'], hue=df_retain['Branch Name']);
plt.xticks(rotation=45);
```

## retain sales branchwise

```
df_retainpie = df_retain.groupby('Branch Name')['Invoice Amount'].sum()
df_retainpie
```

```
df_retainpie.plot.pie(y='Invoice Amount', figsize=(5, 5))
```

Jai 1 has the highest retain sales and Jai-4 has the lowest retain sales.
